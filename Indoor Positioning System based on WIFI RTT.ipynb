```python
import os

path = os.getcwd() + '/RTT-Data-Collection'
files = os.listdir(path)
print(files)
```

```python
import pandas as pd

rtt1_dataframe = pd.read_csv(path + '/EETAC Auditorium-C3-0-google-Pixel 3a-RTT-day 1.csv')
rtt2_dataframe = pd.read_csv(path + '/EETAC Auditorium-C3-0-google-Pixel 3a-RTT-day 2.csv')
rss1_dataframe = pd.read_csv(path + '/EETAC Auditorium-C3-0-google-Pixel 3a-RSS-day 1.csv')
rss2_dataframe = pd.read_csv(path + '/EETAC Auditorium-C3-0-google-Pixel 3a-RSS-day 2.csv')

print(rtt1_dataframe.head())
print(rtt2_dataframe.head())
print(rss1_dataframe.head())
print(rss2_dataframe.head())
```

```python
rtt_dataframe = pd.concat([rtt1_dataframe, rtt2_dataframe], ignore_index=True)
rss_dataframe = pd.concat([rss1_dataframe, rss2_dataframe], ignore_index=True)

print(rtt_dataframe.head())
print(rss_dataframe.head())
```

```python
data_frame = pd.merge(rtt_dataframe, rss_dataframe, on=['batch', 'x', 'y', 'z', 'brand', 'model', 'angle', 'sampleNumber'])
print(data_frame)
```

```python
import numpy as np

data_frame['Y'] = data_frame.apply(lambda row: np.hstack((row.x, row.y, row.z)), axis=1)
print(data_frame['Y'])
```

```python
from sklearn.cluster import KMeans

def cluster(group):
    features = group.iloc[:, 8:-1]

    kmeans = KMeans(n_clusters=5, random_state=0).fit(features)
    group['cluster'] = kmeans.labels_
    return group

data_frame = data_frame.reset_index(drop=True)
data_frame = data_frame.groupby(['x', 'y', 'z'], group_keys=False).apply(cluster)
```
