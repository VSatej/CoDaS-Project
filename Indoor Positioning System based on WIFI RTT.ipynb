```python
basic_columns = ["batch","x","y","z","angle","brand","model","sampleNumber"]


APS = [
"58:cb:52:d7:b5:31",
"58:cb:52:d7:b4:d9",
"58:cb:52:d7:b5:5d",
"58:cb:52:d7:bb:61",
"30:23:03:87:19:6d",
"30:23:03:87:17:99",
"30:23:03:87:17:81"
]
ap_columns = [col for col in data_frame.columns if any(mac in col for mac in APS)]
data_frame = data_frame[basic_columns + ap_columns]
print(data_frame.head())

rtt_columns  = [col for col in data_frame.columns if '#0_samples_averaged' in col]
rss_columns  = [col for col in data_frame.columns if '#1_samples_averaged' in col]
print("RTT columns:", rtt_columns)
print("RSS columns:", rss_columns)

for col in rtt_columns:
    data_frame = data_frame.loc[data_frame[col] == 1]

for col in rss_columns:
    data_frame = data_frame.loc[data_frame[col] == 1]
```

```python
print(data_frame.columns)
data_frame = data_frame.drop(columns=[col for col in data_frame.columns if '#0_samples_averaged' in col or '#1_samples_averaged' in col])
print(data_frame.columns)
```

```python
rss_columns = [col for col in data_frame.columns if '#0_value' in col]
rtt_columns = [col for col in data_frame.columns if '#1_value' in col]
std_columns = [col for col in data_frame.columns if '#1_error' in col]

columns = rss_columns + rtt_columns + std_columns
print("RSS columns:", len(rss_columns))
print("RTT columns:", len(rtt_columns))
print("STD columns:", len(std_columns))
```

```python
from sklearn.cluster import KMeans

def cluster(group):
    features = data_frame.select_dtypes(include=['number'])

    kmeans = KMeans(n_clusters=5, random_state=42).fit(features)
    group['cluster'] = kmeans.labels_
    return group

modified_data_frame = data_frame.reset_index(drop=True)
modified_data_frame = cluster(data_frame)
```

```python
result = (
    modified_data_frame
    .groupby(['x', 'y', 'z'])['cluster']
    .value_counts(normalize=True)
    .mul(100)
    .rename('percentage')
    .reset_index()
)

print(result)
```

```python
rss_columns = [col for col in modified_data_frame.columns if '#0_value' in col]
rtt_columns = [col for col in modified_data_frame.columns if '#1_value' in col]
std_columns = [col for col in modified_data_frame.columns if '#1_error' in col]
```

```python
from sklearn.ensemble import ExtraTreesClassifier

rss_feature_importance_per_cluster = {}

for cluster_id in modified_data_frame['cluster'].unique():
    X = modified_data_frame[rss_columns].select_dtypes(include=['number'])

    y = modified_data_frame['cluster']
    y_binary = (y == cluster_id).astype(int)

    model = ExtraTreesClassifier(n_estimators=100, random_state=42)
    model.fit(X, y_binary)

    importances = model.feature_importances_

    top_idx = importances.argmax()
    top_ap = X.columns[top_idx]
    top_score = importances[top_idx]

    rss_feature_importance_per_cluster[cluster_id] = (top_ap, top_score)
```

```python
rtt_feature_importance_per_cluster = {}

for cluster_id in modified_data_frame['cluster'].unique():
    X = modified_data_frame[rtt_columns].select_dtypes(include=['number'])

    y = modified_data_frame['cluster']
    y_binary = (y == cluster_id).astype(int)

    model = ExtraTreesClassifier(n_estimators=100, random_state=42)
    model.fit(X, y_binary)

    importances = model.feature_importances_

    top_idx = importances.argmax()
    top_ap = X.columns[top_idx]
    top_score = importances[top_idx]

    rtt_feature_importance_per_cluster[cluster_id] = (top_ap, top_score)
```

```python
std_feature_importance_per_cluster = {}

for cluster_id in modified_data_frame['cluster'].unique():
    X = modified_data_frame[std_columns].select_dtypes(include=['number'])

    y = modified_data_frame['cluster']
    y_binary = (y == cluster_id).astype(int)

    model = ExtraTreesClassifier(n_estimators=100, random_state=42)
    model.fit(X, y_binary)

    importances = model.feature_importances_

    top_idx = importances.argmax()
    top_ap = X.columns[top_idx]
    top_score = importances[top_idx]

    std_feature_importance_per_cluster[cluster_id] = (top_ap, top_score)
```

```python
print("Top RSSI features:\n")
for cluster, (ap, score) in rss_feature_importance_per_cluster.items():
    print(f"Cluster {cluster} → Top AP: {ap} (importance: {score:.4f})")

print("\nTop RTT features:\n")
for cluster, (ap, score) in rtt_feature_importance_per_cluster.items():
    print(f"Cluster {cluster} → Top AP: {ap} (importance: {score:.4f})")

print("\nTop STD features:\n")
for cluster, (ap, score) in std_feature_importance_per_cluster.items():
    print(f"Cluster {cluster} → Top AP: {ap} (importance: {score:.4f})")
```

```python
top_features = set()

for cluster, (ap, _) in rss_feature_importance_per_cluster.items():
    top_features.add(ap)

for cluster, (ap, _) in rtt_feature_importance_per_cluster.items():
    top_features.add(ap)

for cluster, (ap, _) in std_feature_importance_per_cluster.items():
    top_features.add(ap)

top_features = list(top_features)
```

```python
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd

candidate_k = [3, 5, 7, 9, 11]

all_rtt = list(rtt_feature_importance_per_cluster.values())

top_2_rtt = [ap for ap, _ in sorted(all_rtt, key=lambda x: x[1], reverse=True)[:2]]

top_2_std = [
    ap.replace('_value', '_error')
    for ap in top_2_rtt
    if ap.replace('_value', '_error') in modified_data_frame.columns
]

selected_features = list(set(top_2_rtt + top_2_std))

X = modified_data_frame[selected_features].select_dtypes(include=['number'])
Y = modified_data_frame[['x', 'y']]

X = X.apply(pd.to_numeric, errors='coerce')

kf = KFold(n_splits=5, shuffle=True, random_state=42)

for k in candidate_k:
    knn = KNeighborsRegressor(n_neighbors=k, weights='distance')

    rmse_scores = []

    for train_idx, test_idx in kf.split(X):
        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = Y.iloc[train_idx], Y.iloc[test_idx]

        scaler = StandardScaler()
        X_train = scaler.fit_transform(X_train)
        X_test = scaler.transform(X_test)

        knn.fit(X_train, y_train)
        y_pred = knn.predict(X_test)

        rmse = np.sqrt(mean_squared_error(y_test, y_pred))
        rmse_scores.append(rmse)

    print(f"k={k}, mean_RMSE={np.mean(rmse_scores):.4f}, std={np.std(rmse_scores):.4f}")
```
